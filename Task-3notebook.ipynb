```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("retail_sales_dataset.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df['Product Category'].value_counts().plot(kind='bar',color = 'c')
plt.title("Product Category Distribution")
plt.xlabel("Product Category")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.groupby('Date')['Total Amount'].sum().plot(color = 'y')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Amount")
plt.show()
```

```python
plt.hist(df['Quantity'],bins = 3,color = 'c',edgecolor = 'black')
plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('Product Category')['Total Amount'].sum().plot(kind = 'bar',color = 'c',width = 0.35)
plt.title("Revenue of Product Category")
plt.xlabel("Product category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
plt.figure(figsize = (6,4))
plt.scatter(df['Price per Unit'], df['Quantity'],color = 'b')
plt.xlabel('Price per Unit')
plt.ylabel('Quantity Purchased')
plt.title('Price per Unit vs Quantity Purchased')
plt.show()
```

```python
sns.boxplot(x='Product Category', y='Total Amount', data=df,color = 'hotpink')
plt.title("Spending Distribution by Category")
plt.show()
```

```python
sns.heatmap(df[['Age','Quantity','Price per Unit','Total Amount']].corr(),annot = True,cmap = 'coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.tight_layout()
plt.show()
```
